```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('Train.csv')
train.head()
```

```python
print(train.duplicated().sum())
```

```python
test = pd.read_csv('Test.csv')
test.head()
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
x_train = train.drop('shares', axis = 1)
y_train = train['shares']
x_test = test
```

```python
from sklearn.ensemble import BaggingRegressor
rf = BaggingRegressor()
rf_pred = rf.fit(x_train, y_train).predict(x_test)
print('Accuracy score', rf.score(x_train, y_train))
solution = pd.DataFrame({'shares': rf_pred})
solution.to_csv('Solution1.csv', index = False)
```

```python
from lightgbm import LGBMRegressor

lgbm = LGBMRegressor(objective = 'regression', boosting = 'dart', metric = 'l2_root', learning_rate =0.01,num_iterations =1000,
         num_leaves = 300, max_depth = -1, min_data_in_leaf = 11, bagging_fraction =0.90, bagging_freq = 2,
         bagging_seed = 3, feature_fraction =  0.90, feature_fraction_seed = 2, max_bin = 200)
y_pred = lgbm.fit(x_train, y_train).predict(x_test)
solution = pd.DataFrame({'tumor_size': y_pred})
solution.to_csv('Solution2.csv', index = False)
```
