```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
tf1 = TfidfVectorizer(min_df=3, token_pattern=r'\w{3,}', ngram_range=(1,3), max_df=0.9)
tf2 = TfidfVectorizer(min_df=2, token_pattern=r'\w{3,}')
tf3 = TfidfVectorizer(min_df=2, token_pattern=r'\w{3,}')
tf4 = TfidfVectorizer(min_df=2, token_pattern=r'\w{3,}')
tf5 = TfidfVectorizer(min_df=2, token_pattern=r'\w{3,}')
tf6 = TfidfVectorizer(min_df=2, token_pattern=r'\w{3,}')
tf7 = TfidfVectorizer(min_df=2, token_pattern=r'\w{3,}')
tf8 = TfidfVectorizer(min_df=2, token_pattern=r'\w{3,}')

X_train_package_name = tf1.fit_transform(X_train_package_name)
X_train_destination = tf2.fit_transform(X_train_destination)
X_train_itinerary = tf3.fit_transform(X_train_itinerary)
X_train_places_covered = tf4.fit_transform(X_train_places_covered)
X_train_hotel_details = tf5.fit_transform(X_train_hotel_details)
X_train_airline = tf6.fit_transform(X_train_airline)
X_train_sightseeing = tf7.fit_transform(X_train_sightseeing)
X_train_cancellation = tf8.fit_transform(X_train_cancellation)

X_cv_package_name = tf1.transform(X_cv_package_name)
X_cv_destination = tf2.transform(X_cv_destination)
X_cv_itinerary = tf3.transform(X_cv_itinerary)
X_cv_places_covered = tf4.transform(X_cv_places_covered)
X_cv_hotel_details = tf5.transform(X_cv_hotel_details)
X_cv_airline = tf6.transform(X_cv_airline)
X_cv_sightseeing = tf7.transform(X_cv_sightseeing)
X_cv_cancellation = tf8.transform(X_cv_cancellation)
```

```python
from scipy import sparse
from sklearn.preprocessing import StandardScaler

sc1 = StandardScaler()
X_train_flight_stops = sc1.fit_transform(np.array(X_train['Flight Stops']).reshape(-1,1))
X_cv_flight_stops = sc1.transform(np.array(X_cv['Flight Stops']).reshape(-1,1))
X_train_flight_stops = sparse.csr_matrix(X_train_flight_stops)
X_cv_flight_stops = sparse.csr_matrix(X_cv_flight_stops)

sc2 = StandardScaler()
X_train_meal = sc2.fit_transform(np.array(X_train['Meals']).reshape(-1,1))
X_cv_meal = sc2.transform(np.array(X_cv['Meals']).reshape(-1,1))
X_train_meal = sparse.csr_matrix(X_train_meal)
X_cv_meal = sparse.csr_matrix(X_cv_meal)

sc3 = StandardScaler()
X_train_package_type = sc3.fit_transform(np.array(X_train['Package Type']).reshape(-1,1))
X_cv_package_type = sc3.transform(np.array(X_cv['Package Type']).reshape(-1,1))
X_train_package_type = sparse.csr_matrix(X_train_package_type)
X_cv_package_type = sparse.csr_matrix(X_cv_package_type)

sc4 = StandardScaler()
X_train_start_city = sc4.fit_transform(np.array(X_train['Start City']).reshape(-1,1))
X_cv_start_city = sc4.transform(np.array(X_cv['Start City']).reshape(-1,1))
X_train_start_city = sparse.csr_matrix(X_train_start_city)
X_cv_start_city = sparse.csr_matrix(X_cv_start_city)
```

```python
from scipy.sparse import hstack, csr_matrix
```

```python
merged_train = hstack((X_train_airline, X_train_cancellation, X_train_destination, X_train_flight_stops, X_train_hotel_details, X_train_itinerary, X_train_meal, X_train_package_name, X_train_package_type, X_train_places_covered, X_train_sightseeing, X_train_start_city))
merged_cv  = hstack((X_cv_airline, X_cv_cancellation, X_cv_destination, X_cv_flight_stops, X_cv_hotel_details, X_cv_itinerary, X_cv_meal, X_cv_package_name, X_cv_package_type, X_cv_places_covered, X_cv_sightseeing, X_cv_start_city))
```

```python
merged_train.shape, merged_cv.shape
```

```python
import lightgbm as lgb
train_data = lgb.Dataset(merged_train, label=y_train)
test_data = lgb.Dataset(merged_cv, label=y_cv)
```

```python
param = {'objective': 'regression',
         'boosting': 'gbdt',  
         'metric': 'l2_root',
         'learning_rate': 0.01, 
         'num_iterations': 3500,
         'num_leaves': 31,
         'max_depth': -1,
         'min_data_in_leaf': 11,
         'bagging_fraction': 0.90,
         'bagging_freq': 2,
         'bagging_seed': 3,
         'feature_fraction': 0.90,
         'feature_fraction_seed': 2,
         'early_stopping_round': 200,
         'max_bin': 1000
         }
```

```python
lgbm = lgb.train(params=param, train_set=train_data, num_boost_round=100, valid_sets=[test_data])
y_pred_class = lgbm.predict(merged_cv)
```

```python
from sklearn.metrics import mean_squared_error
predictions = []
for x in y_pred_class:
    predictions.append(np.argmax(x))

print('accuracy:', np.sqrt(mean_squared_error(y_cv, predictions)))
```
