```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_excel('Data_Train.xlsx')
train.head()
```

```python
train.shape
```

```python
test = pd.read_excel('Test_set.xlsx')
test.head()
```

```python
test.shape
```

```python
df = pd.concat([train, test], axis = 0, sort = False)
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = df['Date_of_Journey'].apply(lambda x:x.split(' ')[0])
```

```python
df['Month'] = df['Date_of_Journey'].apply(lambda x:x.split(' ')[1])
```

```python
df['Year'] = df['Date_of_Journey'].apply(lambda x:x.split(' ')[2])
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Total_Stops'].value_counts()
```

```python
df['Total_Stops'] = df['Total_Stops'].fillna('1 stop')
```

```python
df['Total_Stops'] = df['Total_Stops'].replace({'non-stop': '0 stop'})
```

```python
df['Total_Stops'] = df['Total_Stops'].apply(lambda x:x.split(' ')[0])
df['Total_Stops'].value_counts()
```

```python
df['Total_Stops'] = df['Total_Stops'].astype(int)
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].apply(lambda x:x.split(':')[0])
```

```python
df['Arrival_Minute'] = df['Arrival_Time'].apply(lambda x:x.split(':')[1])
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)
```

```python
df['Departure_Hour'] = df['Dep_Time'].apply(lambda x:x.split(':')[0])
```

```python
df['Departure_Minute'] = df['Dep_Time'].apply(lambda x:x.split(':')[1])
```

```python
df['Departure_Hour'] = df['Departure_Hour'].astype(int)
df['Departure_Minute'] = df['Departure_Minute'].astype(int)
```

```python
df['Route'].value_counts()
```

```python
df['Route1'] = df['Route'].str.split('→ ').str[0]
df['Route2'] = df['Route'].str.split('→ ').str[1]
df['Route3'] = df['Route'].str.split('→ ').str[2]
df['Route4'] = df['Route'].str.split('→ ').str[3]
df['Route5'] = df['Route'].str.split('→ ').str[4]
```

```python
df['Route1'] = df['Route1'].fillna('None')
df['Route2'] = df['Route2'].fillna('None')
df['Route3'] = df['Route3'].fillna('None')
df['Route4'] = df['Route4'].fillna('None')
df['Route5'] = df['Route5'].fillna('None')
```

```python
df = df.drop(['Dep_Time', 'Arrival_Time', 'Date_of_Journey', 'Route', 'Duration'], axis = 1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df["Airline"] = le.fit_transform(df['Airline'])
df["Source"] = le.fit_transform(df['Source'])
df["Destination"] = le.fit_transform(df['Destination'])
df["Additional_Info"] = le.fit_transform(df['Additional_Info'])
df['Route1'] = le.fit_transform(df['Route1'])
df['Route2'] = le.fit_transform(df['Route2'])
df['Route3'] = le.fit_transform(df['Route3'])
df['Route4'] = le.fit_transform(df['Route4'])
df['Route5'] = le.fit_transform(df['Route5'])
```

```python
newtrain = df[0:10683]
newtest = df[10683:]
```

```python
newtest = newtest.drop('Price', axis = 1)
```

```python
x_train = newtrain.drop('Price', axis = 1)
y_train = newtrain['Price']
x_test = newtest.copy()
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators, 'max_features': max_features, 'max_depth': max_depth,
               'min_samples_split': min_samples_split, 'min_samples_leaf': min_samples_leaf}
print(random_grid)
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 50, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
y_pred = rf_random.fit(x_train, y_train).predict(x_test)
```

```python
solution = pd.DataFrame({'Price': np.abs(y_pred)})
solution.to_excel('Solution2.xlsx', index = False)
```
